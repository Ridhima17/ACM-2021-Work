```python
#Question 1 Menu driven program

def add(a,b):
    c = a + b
    print("Addition result: ",c)
    
def subtract(a,b):
    c = a - b
    print("Subtraction result: ",c)

def multiply(a,b):
    c = a * b
    print("Multiplication result: ",c)

def divide(a,b):
    try:
        c = a/b
        print("Division result: ",c)
    except:
        print("Can't divide by zero")

while(True):
    a = input("Enter the first number")
    b = input("Enter the second number")
    print("Main Menu")
    print("1. ADD")
    print("2. SUBTRACT")
    print("3. MULTIPLY")
    print("4. DIVIDE")
    print("5. EXIT")
    n = input("Enter your choice: ")
    if(n == 1):
        add(a,b)
    elif(n == 2):
        subtract(a,b)
    elif(n = 3):
        multiply(a,b)
    elif(n == 4):
        divide(a,b)
    elif(n == 5):
        break
    else:
        print("WRONG CHOICE")
    
```

```python
#Question 2 - Reverse a sentence using list and string data structures
sentence = input("Enter the sentence: ")
word_list = sentence.split()
reversed_list = word_list[::-1]
reversed_sentence = " ".join(reversed_list)
print(reversed_sentence)
```

```python
#Question 3 - Counting number of characters from a text file
f = open("myfile.txt","w")
L = ["This is Delhi \n","I live in Ghaziabad \n", "I like Paris \n", "Tokyo is also nice though \n","But anything just can't beat Maldives \n"]
f.writelines(L)
f.close()

f1 = open("myfile.txt","r")
all_freq = {}
for i in f1.read():
        all_freq[i] = all_freq.get(i,0) + 1
print("Count of alphabets present in the text file: ", str(all_freq))
```
