```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
colored_img = cv2.imread('MTlab3.png',1)

grayscale_img = cv2.imread('MTlab3.png',0)
```

```python
colored_dim = colored_img.shape
print('Dimension: ',colored_dim)
grayscale_dim = grayscale_img.shape
print('Dimension: ',grayscale_dim)
```

```python
cv2_imshow(colored_img)
```

```python
cv2_imshow(grayscale_img)
```

```python
# Image downscaling
down_height = 200
down_width = 400
down_points = (down_width,down_height)
resized_down = cv2.resize(colored_img,down_points)
dimension_downscaling = resized_down.shape
print('Dimensions after downscaling: ', dimension_downscaling);
cv2_imshow(resized_down)
```

```python
# Image upscaling
up_height = 470
up_width = 670
up_points = (up_height,up_width)
resized_up = cv2.resize(colored_img,up_points)
dimension_upscaling = resized_up.shape
print('Dimensions after upscaling: ', dimension_upscaling)
cv2_imshow(resized_up)
```

```python
# Image cropping
cropped_image = colored_img[100:200 , 100:200]
cropped_dimension = cropped_image.shape
print('Dimensions of the cropped image: ',cropped_dimension)
cv2_imshow(cropped_image)
```

```python
# Image rotation
height, width = colored_img.shape[:2]
center = (width/2,height/2)
rotate_matrix = cv2.getRotationMatrix2D(center=center,angle=30,scale=1)
rotated_image = cv2.warpAffine(src=colored_img,M=rotate_matrix,dsize=(width,height))

cv2_imshow(rotated_image)
```
